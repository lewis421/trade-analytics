```python
# Define the ticker symbol for Apple
import yfinance as yf
ticker_symbol = 'AAPL'

# Define the start and end dates for the data
start_date = '2020-01-01'  # Example start date
end_date = '2025-01-01'  # Example end date (Note: the current date is August 16, 2025, so adjust the end date for more recent data if desired)

# Download the data
df = yf.download(ticker_symbol, start=start_date, end=end_date)
df.to_csv("apple_daily_ohlcv.csv")

```
